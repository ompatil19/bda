```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = './birthday.csv'
data = pd.read_csv(file_path)

month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
data['month'] = data['month'].map(month_names)

plt.figure(figsize=(10, 7))
plt.pie(data['users'], labels=data['month'], autopct='%1.1f%%', startangle=140)
plt.title('Users\' Birthdays by Month')
plt.axis('equal')  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = './age.csv'
data = pd.read_csv(file_path)

total_users = data['total_users'][0]
users_above_25 = data['users_above_25'][0]
users_below_25 = total_users - users_above_25

labels = ['Users above 25', 'Users below 25']
sizes = [users_above_25, users_below_25]

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('User Age Distribution')
plt.axis('equal')  
plt.show()
```
